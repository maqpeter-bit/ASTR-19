```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, np.pi, 100)
#^ Defining an array here to be used as X
#print(x)
```

```python
def journalFunction(a, b):
    return a * np.exp(-b*x)
```

```python
a = 3
b = 4

y = journalFunction(a, b)

```

```python
plt.plot(x)
plt.plot(y)
plt.xlabel("x label")
plt.ylabel("y label")

plt.savefig("CodingJournal#2 Session1 Plot")
```
